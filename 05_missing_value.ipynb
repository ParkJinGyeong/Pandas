### fillna
- 원본 자체가 바뀌지 않음 

```python
df = sns.load_dataset('titanic')
```

```python
df1 = df.copy() #카피본 만드는 것 추천 ! 
```

```python
df1['age'].fillna(100).tail()
```

```python
df1['age'] = df1['age'].fillna(100) #원본을 직접적으로 수정해주는 것 
```

```python
df1.tail() 
```

```python
df['deck'].fillna("C") 
```

```python
# df['deck'].fillna("no data") # 카테고리타입은 카테고리에 없는 정보를 넣을 수 없음 
```

```python
df1['deck'].cat.add_categories('no data')
#그렇다면 카테고리를 직접 추가해서 채워주면 됨
```

```python
# df1['deck'].fillna("no data") # 엥 왜 에러가 뜨지 ? 
```

```python
df1 = df.copy()
```

```python
df1['age']
```

```python
mean_age = df1['age'].mean()
df1['age'].fillna(mean_age).tail()
```

```python
median_age = df1['age'].median()
df1['age'].fillna(median_age).tail()
```

- 남자 승객의 결측치 => 남자 승객의 평균
- 여자 승객의 결측치 => 여자 승객의 평균 

```python
# c1 = df[(df['sex'] == 'male') & (df['age'])] 왜 안 되는지 확인 필요
```

```python
#cond1 = df['sex'] == 'male'
#cond2 = df['age']

#df.loc[cond1 & cond2]
#왜안되는지 확인 필요
```

```python
c1 = df['sex'] == 'male'
male_mean = df.loc[c1, 'age'].mean()
df1.loc[c1,'age'] = df1.loc[c1, 'age'].fillna(male_mean)

c2 = df['sex'] == 'female'
female_mean = df.loc[c2, 'age'].mean()
df1.loc[c2,'age'] = df1.loc[c2, 'age'].fillna(female_mean)
```

```python
df1['age'].mean()
```

```python
df1['deck'].mode()
```

```python
df1['deck'].fillna(df1['deck'].mode()[0]).head()
#mode()[0] 해야하는 이유 
```
