```python
m = df['embarked'].mode()[0]
df['embarked'] = df['embarked'].fillna(m)
```

```python
assert 0 == df['embarked'].isnull().sum()
```

```python
def nan(x):
    mean = x.mean()
    return x.fillna(mean)

df['age'] = df.groupby('sex', group_keys=False)['age'].apply(nan)
df.tail()
```

```python

df.groupby('sex', group_keys=False)['age'].mean()
```

```python
assert df.groupby('sex')['age'].mean()['female'].round(4) == 27.9157
assert df.groupby('sex')['age'].mean()['male'].round(4) == 30.7266
assert df['age'].isnull().sum() == 0
```
